```python
# ============================================================================
# CELL 1: K·∫æT N·ªêI MONGODB V√Ä THI·∫æT L·∫¨P BAN ƒê·∫¶U
# ============================================================================

import pymongo
from pprint import pprint
import pandas as pd

print("üîå ƒêang k·∫øt n·ªëi ƒë·∫øn MongoDB...")

try:
    # K·∫øt n·ªëi ƒë·∫øn MongoDB (gi·ªëng connection string c·ªßa Compass)
    client = pymongo.MongoClient("mongodb://localhost:27017/")
    
    # Ki·ªÉm tra k·∫øt n·ªëi
    client.admin.command('ping')
    print("‚úÖ K·∫øt n·ªëi MongoDB th√†nh c√¥ng!")
    
    # T·∫°o database
    db = client["document_db"]
    
    # Cleanup collections c≈©
    db.subjects.drop()
    db.objects.drop() 
    db.policies.drop()
    
    print("‚úÖ ƒê√£ x√≥a collections c≈©")
    
except Exception as e:
    print(f"‚ùå L·ªói k·∫øt n·ªëi MongoDB: {e}")
    raise
```

```python
# ============================================================================
# CELL 2: INSERT T·∫§T C·∫¢ D·ªÆ LI·ªÜU (SUBJECTS, OBJECTS, POLICIES)
# ============================================================================

print("üë• ƒêang insert d·ªØ li·ªáu Subjects...")

subjects_data = [
    {
        "_id": "subject_alice",
        "hierarchy": {
            "organization": "ABC_Corp",
            "company": "ABC_Inc",
            "division": "technology", 
            "department": "engineering",
            "branch": "headquarters",
            "group": "dev_team"
        },
        "attributes": {
            "user_id": "user1",
            "user_name": "alice",
            "role": "developer"
        }
    },
    {
        "_id": "subject_eve", 
        "hierarchy": {
            "organization": "FinanceCorp",
            "company": "Finance_Inc",
            "division": "operations",
            "department": "finance", 
            "branch": "main_office",
            "group": "payroll_team"
        },
        "attributes": {
            "user_id": "user2",
            "user_name": "eve",
            "role": "specialist"
        }
    }
]

result_subjects = db.subjects.insert_many(subjects_data)
print(f"‚úÖ ƒê√£ insert {len(result_subjects.inserted_ids)} subjects")

# Hi·ªÉn th·ªã subjects ƒë√£ insert
print("\nüìã DANH S√ÅCH SUBJECTS:")
for subject in db.subjects.find():
    pprint(subject)

# ============================================================================
# INSERT D·ªÆ LI·ªÜU OBJECTS (T√ÄI NGUY√äN)
# ============================================================================

print("üìÅ ƒêang insert d·ªØ li·ªáu Objects...")

objects_data = [
    {
        "_id": "object_spec",
        "hierarchy": {
            "organization": "ABC_Corp",
            "company": "ABC_Inc", 
            "division": "technology",
            "department": "engineering",
            "branch": "headquarters",
            "project": "product_x"
        },
        "attributes": {
            "file_id": "file1",
            "file_name": "product_spec.pdf",
            "classification": "confidential"
        }
    },
    {
        "_id": "object_financial",
        "hierarchy": {
            "organization": "FinanceCorp",
            "company": "Finance_Inc",
            "division": "operations", 
            "department": "finance",
            "branch": "main_office", 
            "project": "annual_reports"
        },
        "attributes": {
            "file_id": "file2",
            "file_name": "financial_report.pdf", 
            "classification": "secret"
        }
    },
    {
        "_id": "object_technical_guide",
        "hierarchy": {
            "organization": "ABC_Corp",
            "company": "ABC_Inc", 
            "division": "technology",
            "department": "engineering",
            "branch": "headquarters",
            "project": "product_x"
        },
        "attributes": {
            "file_id": "file3",
            "file_name": "technical_guide.pdf",
            "classification": "internal",
        }
    }
]

result_objects = db.objects.insert_many(objects_data)
print(f"‚úÖ ƒê√£ insert {len(result_objects.inserted_ids)} objects")

# Hi·ªÉn th·ªã objects ƒë√£ insert
print("\nüìã DANH S√ÅCH OBJECTS:")
for obj in db.objects.find():
    pprint(obj)

# ============================================================================
# INSERT D·ªÆ LI·ªÜU POLICIES (CH√çNH S√ÅCH)
# ============================================================================

print("üîê ƒêang insert d·ªØ li·ªáu Policies...")

policies_data = [
    {
        "policy_id": "user_level",
        "subject": { 
            "hierarchy": { 
                "company": "ABC_Inc",
                "division": "technology",
                "department": "engineering", 
                "branch": "headquarters",
                "group": "dev_team"
            },
            "attributes": { "user_name": "alice" }
        },
        "object": {
            "hierarchy": { 
                "company": "ABC_Inc",
                "division": "technology",
                "department": "engineering",
                "branch": "headquarters" 
            },
            "attributes": { "file_name": "product_spec.pdf" }
        },
        "action": "read",
        "permission": "allow"
    },
    {
        "policy_id": "group_level",
        "subject": {
            "hierarchy": { 
                "company": "ABC_Inc",
                "division": "technology",
                "department": "engineering",
                "branch": "headquarters",
                "group": "dev_team"
            }
        },
        "object": {
            "hierarchy": { 
                "company": "ABC_Inc",
                "division": "technology", 
                "department": "engineering",
                "branch": "headquarters"
            }
        },
        "action": "write", 
        "permission": "allow"
    },
    {
        "policy_id": "dept_level",
        "subject": {
            "hierarchy": { 
                "company": "ABC_Inc",
                "division": "technology",
                "department": "engineering"
            }
        },
        "object": {
            "hierarchy": { 
                "company": "ABC_Inc",
                "division": "technology",
                "department": "engineering"
            }
        },
        "action": "read",
        "permission": "allow" 
    },
    {
        "policy_id": "branch_level", 
        "subject": {
            "hierarchy": {
                "company": "ABC_Inc",
                "division": "technology", 
                "branch": "headquarters"
            }
        },
        "object": {
            "hierarchy": {
                "company": "ABC_Inc",
                "division": "technology",
                "branch": "headquarters"
            }
        },
        "action": "read",
        "permission": "allow"
    },
    {
        "policy_id": "division_level",
        "subject": {
            "hierarchy": {
                "company": "ABC_Inc",
                "division": "technology"
            }
        },
        "object": {
            "hierarchy": {
                "company": "ABC_Inc", 
                "division": "technology"
            }
        },
        "action": "read",
        "permission": "allow"
    },
    {
        "policy_id": "cross_org_deny",
        "subject": {
            "hierarchy": { "organization": "FinanceCorp" }
        },
        "object": {
            "hierarchy": { "organization": "ABC_Corp" }
        },
        "action": "read",
        "permission": "deny"
    }
]

result_policies = db.policies.insert_many(policies_data)
print(f"‚úÖ ƒê√£ insert {len(result_policies.inserted_ids)} policies")

# Hi·ªÉn th·ªã policies ƒë√£ insert
print("\nüìã DANH S√ÅCH POLICIES:")
for policy in db.policies.find():
    pprint(policy)
```

```python
# ============================================================================
# CELL 3: QUERY 1 - USER LEVEL
# Query 1: L·ªçc theo N√∫t L√° (Level 5 Attribute) c·ªßa Ch·ªß th·ªÉ
# M·ª•c ti√™u: T√¨m t·∫•t c·∫£ c√°c ch√≠nh s√°ch √°p d·ª•ng cho ng∆∞·ªùi d√πng c·ª• th·ªÉ l√† 'alice'.
# ============================================================================

print("=" * 80)
print("1. USER LEVEL: Alice c√≥ th·ªÉ ƒë·ªçc file product_spec.pdf kh√¥ng?")
print("=" * 80)

user_query = db.policies.find_one({
    "subject.attributes.user_name": "alice",
    "object.attributes.file_name": "product_spec.pdf"
})

if user_query:
    print("‚úÖ K·∫æT QU·∫¢: Alice C√ì TH·ªÇ ƒë·ªçc product_spec.pdf")
    print("üìÑ POLICY DOCUMENT:")
    pprint(user_query)
else:
    print("‚ùå K·∫æT QU·∫¢: Alice KH√îNG TH·ªÇ ƒë·ªçc product_spec.pdf")
```

```python
# ============================================================================
# CELL 4: QUERY 2 - GROUP LEVEL  
# ============================================================================

print("\n" + "=" * 80)
print("2. GROUP LEVEL: Nh√≥m dev_team c√≥ th·ªÉ thao t√°c nh·ªØng g√¨ trong d·ª± √°n?")
print("=" * 80)

group_query = db.policies.find_one({
    "subject.hierarchy.group": "dev_team"
})

if group_query:
    action_text = "GHI v√† C·∫¨P NH·∫¨T" if group_query["action"] == "write" else "ƒê·ªåC"
    print(f"‚úÖ K·∫æT QU·∫¢: dev_team c√≥ th·ªÉ {action_text} t√†i li·ªáu d·ª± √°n")
    print("üìÑ POLICY DOCUMENT:")
    pprint(group_query)
else:
    print("‚ùå K·∫æT QU·∫¢: Kh√¥ng t√¨m th·∫•y quy·ªÅn c·ª• th·ªÉ cho dev_team")
```

```python
# ============================================================================
# CELL 5: QUERY 3 - DEPARTMENT LEVEL
# ============================================================================

print("\n" + "=" * 80)
print("3. DEPARTMENT LEVEL: Ph√≤ng Engineering ƒë∆∞·ª£c ph√©p truy c·∫≠p t√†i nguy√™n n√†o?")
print("=" * 80)

dept_policies = list(db.policies.find({
    "subject.hierarchy.department": "engineering"
}))

if dept_policies:
    print("‚úÖ K·∫æT QU·∫¢: Ph√≤ng Engineering c√≥ quy·ªÅn truy c·∫≠p c√°c file sau:")
    
    # L·∫•y danh s√°ch c√°c file ƒë∆∞·ª£c access t·ª´ policies
    accessible_files = []
    for policy in dept_policies:
        if (policy.get("object") and 
            policy["object"].get("attributes") and 
            policy["object"]["attributes"].get("file_name")):
            accessible_files.append({
                "file_name": policy["object"]["attributes"]["file_name"],
                "action": policy["action"],
                "permission": policy["permission"]
            })
    
    # Hi·ªÉn th·ªã file c√≥ th·ªÉ truy c·∫≠p
    if accessible_files:
        print("üìÅ C√°c file c√≥ th·ªÉ truy c·∫≠p:")
        for i, file in enumerate(accessible_files[:2], 1):
            print(f"   {i}. {file['file_name']} - Quy·ªÅn: {file['action'].upper()} - {file['permission'].upper()}")
        
        if len(accessible_files) > 2:
            print(f"   ... v√† {len(accessible_files) - 2} file kh√°c")
    
    # Hi·ªÉn th·ªã policy document ƒë·∫ßu ti√™n
    print("\nüìÑ POLICY DOCUMENT:")
    pprint(dept_policies[0] if dept_policies else {})
    
else:
    print("‚ùå K·∫æT QU·∫¢: Kh√¥ng t√¨m th·∫•y policy cho ph√≤ng Engineering")
```

```python
# ============================================================================
# CELL 6: QUERY 4 - BRANCH LEVEL
# ============================================================================

print("\n" + "=" * 80)
print("4. BRANCH LEVEL: Chi nh√°nh headquarters c√≥ nh·ªØng quy·ªÅn g√¨?")
print("=" * 80)

branch_query = db.policies.find_one({
    "subject.hierarchy.branch": "headquarters"
})

if branch_query:
    privilege_text = "QU·∫¢N L√ù v√† CH·ªàNH S·ª¨A" if branch_query["action"] == "write" else "TI·∫æP C·∫¨N v√† XEM"
    print(f"‚úÖ K·∫æT QU·∫¢: Chi nh√°nh headquarters c√≥ ƒë·∫∑c quy·ªÅn {privilege_text} d·ªØ li·ªáu")
    print("üìÑ POLICY DOCUMENT:")
    pprint(branch_query)
else:
    print("‚ùå K·∫æT QU·∫¢: Kh√¥ng t√¨m th·∫•y ƒë·∫∑c quy·ªÅn cho chi nh√°nh headquarters")
```

```python
# ============================================================================
# CELL 7: QUERY 5 - ORGANIZATION LEVEL
# ============================================================================

print("\n" + "=" * 80)
print("5. ORGANIZATION LEVEL: Cross-organization rules")
print("=" * 80)

org_query = db.policies.find_one({
    "subject.hierarchy.organization": "FinanceCorp"
})

if org_query:
    print(f"‚úÖ K·∫æT QU·∫¢: FinanceCorp organization c√≥ quy·ªÅn {org_query['action']} - {org_query['permission']}")
    print("üìÑ POLICY DOCUMENT:")
    pprint(org_query)
else:
    print("‚ùå K·∫æT QU·∫¢: Kh√¥ng t√¨m th·∫•y policy cho FinanceCorp organization")
```

```python
# ============================================================================
# CELL 8: QUERY 6 - EVE ACCESS CHECK
# ============================================================================

print("\n" + "=" * 80)
print("6. EVE ACCESS CHECK: Eve c√≥ th·ªÉ truy c·∫≠p ABC_Inc kh√¥ng?")
print("=" * 80)

eve_access_query = db.policies.find_one({
    "$or": [
        { "subject.attributes.user_name": "eve", "object.hierarchy.company": "ABC_Inc" },
        { "subject.hierarchy.organization": "FinanceCorp", "object.hierarchy.organization": "ABC_Corp" }
    ]
})

if eve_access_query:
    if eve_access_query["permission"] == "allow":
        print("‚úÖ K·∫æT QU·∫¢: Eve C√ì TH·ªÇ truy c·∫≠p ABC_Inc")
    else:
        print("‚ùå K·∫æT QU·∫¢: Eve KH√îNG TH·ªÇ truy c·∫≠p ABC_Inc")
    print("üìÑ POLICY DOCUMENT:")
    pprint(eve_access_query)
else:
    print("‚ùå K·∫æT QU·∫¢: Eve KH√îNG TH·ªÇ truy c·∫≠p ABC_Inc (kh√¥ng c√≥ policy)")
```
